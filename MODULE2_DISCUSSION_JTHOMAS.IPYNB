```python
import matplotlib.pyplot as plt

# Generate data using a list comprehension
numbers = [x for x in range(1,11)]

plt.plot(numbers)
plt.show()
```

```python
# Generate two datasets 
x_values = [x for x in range(1, 11)]
y_values1 = [x ** 2 for x in x_values]
y_values2 = [x ** 1.5 for x in x_values]

plt.plot(x_values, y_values1, label="x squared", marker="o", color="blue")
plt.plot(x_values, y_values2, label="x to the 1.5 power", marker="s", color="green")


plt.title("Comparison of Two Mathematical Functions")
plt.xlabel("x Values")
plt.ylabel("Y Values")
plt.legend()


plt.show()
```
