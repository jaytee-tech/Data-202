## Part 2 — Basic Queries with Filters and Calculations



### Problem 1 — Employee tenure

**Goal:** Find employees who have been with the company more than **20 years**.

**Requirements:**  
- Use the `Employee` table.  
- Return columns: `FirstName`, `LastName`, and a calculated column `YearsWorked`.  
- You may use SQLite date functions (e.g., `strftime`) to calculate years.

**Task:** Write a SQL query in the string below.


```python

query = """
-- TODO: Employees with more than 20 years at the company.
-- Return FirstName, LastName, and YearsWorked.
-- Hint: Use HireDate and date functions.
"""
query = """
SELECT 
    FirstName,
    LastName,
    CAST((julianday('now') - julianday(date(HireDate))) / 365.25 AS INTEGER) AS YearsWorked
FROM employees
WHERE (julianday('now') - julianday(date(HireDate))) / 365.25 > 20;
"""

run_query(query)
run_query(query)
```


### Problem 2 — Invoice activity by day

**Goal:** Understand which days had the most invoice activity.

**Requirements:**  
- Use the `Invoice` table.  
- Return:  
  - `InvoiceDate` (date only, no time)  
  - `NumInvoices` (number of invoices on that date)  
  - `PctOfTotal` (percentage of all invoices that occurred on that date)  
- Sort by `NumInvoices` in **descending** order.

**Hints:**  
- Use `DATE(InvoiceDate)` to extract the date portion.  
- Use a subquery to get the total number of invoices.  
- Multiply by `100.0` to compute a percentage.


```python

query = """
-- TODO: Invoice activity by date with percentage of total invoices.
"""
query = """
SELECT 
    InvoiceDate,
    COUNT(*) AS NumInvoices,
    ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM invoices), 2) AS PercentOfTotal
FROM invoices
GROUP BY InvoiceDate
ORDER BY InvoiceDate;
"""

run_query(query)
```


**Question:** Which dates show particularly high invoice activity? What might this suggest about customer behavior or store operations?


The highest invoice activiy occurs on several dates 2 invoices were recorded in a single day. Since most dates only contain 1 invoice, these dates represent the peak transaction volume in the dataset. This suggests that customer purchases are generally evenly distributed, with occasional small spikes rather than large surges.

## Part 3 — Joins and Grouping



### Problem 3 — Genres and track counts

**Goal:** Return the name of each **genre** and the number of tracks in that genre.

**Requirements:**  
- Use the `Genre` and `Track` tables.  
- Join them on their key relationship.  
- Return:  
  - `GenreName`  
  - `NumTracks`  
- Sort the results by `NumTracks` in **descending** order.


```python

query = """
-- TODO: Genre name and number of tracks per genre.
"""
query = """
SELECT 
    g.Name AS Genre,
    COUNT(t.TrackId) AS NumTracks
FROM tracks t
JOIN genres g ON t.GenreId = g.GenreId
GROUP BY g.Name
ORDER BY NumTracks DESC;
"""

run_query(query)
```


### Problem 4 — Customers with purchases

**Goal:** Find all customers who have made at least one purchase.

**Requirements:**  
- Use the `Customer` and `Invoice` tables.  
- Return:  
  - `FirstName`, `LastName`  
  - `NumInvoices` (how many invoices each customer has)  
- Show only customers with at least one invoice.  
- Sort by `NumInvoices` in **descending** order.


```python

query = """
-- TODO: Customers with at least one invoice and their invoice count.
"""
query = """
SELECT 
    c.FirstName,
    c.LastName,
    COUNT(i.InvoiceId) AS InvoiceCount
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
GROUP BY c.CustomerId
HAVING COUNT(i.InvoiceId) >= 1
ORDER BY InvoiceCount DESC;
"""

run_query(query)
```


**Question:** Which customers appear to be your most active? How might this information be used in a real business context?


The most active customers each have 7 invoices, which consist of people like Luis Goncalves, and Leonie Kohler. Multiple customers tying for the highest value suggests that purchasing behavior is fairly balanced among top customers.

## Part 4 — Playlists and Rock Tracks (Intermediate)

### Problem 5 — Rock-heavy playlists

**Goal:** Understand which playlists are dominated by **Rock** tracks.

**Requirements:**  
For each playlist, return:
- `PlaylistName`  
- `TotalTracks` (total number of tracks in the playlist)  
- `RockTracks` (number of tracks in the playlist that belong to the Rock genre)  
- `PctRock` (percentage of tracks in the playlist that are Rock)  

Sort the results by `PctRock` in **descending** order.

**Hints:**  
- You will need several tables: `Playlist`, `PlaylistTrack`, `Track`, and `Genre`.  
- Filter genres where the name is `'Rock'`.  
- Use aggregate functions and `GROUP BY` on playlist.


```python

query = """
-- TODO: Playlist stats showing total tracks, rock tracks, and percentage rock.
"""
query = """
SELECT 
    p.Name AS Playlist,
    COUNT(pt.TrackId) AS TotalTracks,
    SUM(CASE WHEN g.Name = 'Rock' THEN 1 ELSE 0 END) AS RockTracks,
    ROUND(
        100.0 * SUM(CASE WHEN g.Name = 'Rock' THEN 1 ELSE 0 END) / COUNT(pt.TrackId),
        2
    ) AS PercentRock
FROM playlists p
JOIN playlist_track pt ON p.PlaylistId = pt.PlaylistId
JOIN tracks t ON pt.TrackId = t.TrackId
JOIN genres g ON t.GenreId = g.GenreId
GROUP BY p.PlaylistId
ORDER BY PercentRock DESC;
"""

run_query(query)
```


**Question:** Which playlist is the most “Rock-heavy”? Does that outcome align with what you would expect from a digital media store?


The Grugne playlist is the most "Rock-heavy" with a total of 15 tracks 14 of them are Rock tracks. This is the highest Rock concentration among playlists. The result is reasonable because Grunge is a subgenre strongly associated with Rock music which should be the main focus of a digital media store to capture.
