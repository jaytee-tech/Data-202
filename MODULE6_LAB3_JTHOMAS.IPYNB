In this lab, I observed that using an indexed lookup with `.loc[]` was noticeably faster than using Boolean filtering, especially when retrieving a specific student by ID. The Boolean filter had to scan the entire column to find a match, while the indexed lookup accessed the row directly, which makes it more efficient as the dataset grows. I also learned that resetting an index is helpful when I need the indexed column back as a regular column for merging, exporting, or performing other operations that don’t require optimized lookups. Overall, this lab helped me understand how Pandas indexing is similar to SQL indexes, since both improve query performance by organizing data for faster retrieval, and how these same ideas apply to larger distributed systems where efficient data access is critical for handling massive datasets.
